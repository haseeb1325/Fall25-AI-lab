```python



df = df.drop_duplicates()


df = df.reset_index(drop=True)


print("Data Pre-Processing Done!")
print(df.head())
```

```python



print("Null Values Before:")
print(df.isnull().sum())


df[df.select_dtypes(include=['float', 'int']).columns] = \
df.select_dtypes(include=['float', 'int']).fillna(df.mean(numeric_only=True))


df[df.select_dtypes(include=['object']).columns] = \
df.select_dtypes(include=['object']).fillna(df.mode().iloc[0])


print("\nNull Values After:")
print(df.isnull().sum())
```

```python



for col in df.columns:
    if df[col].dtype == 'float' or df[col].dtype == 'int':
        if (df[col] % 1 == 0).all():        
            df[col] = df[col].astype(int)


print("Updated Data Types:")
print(df.dtypes)
```
